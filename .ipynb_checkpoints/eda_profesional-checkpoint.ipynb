```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style("whitegrid")

# Crear carpeta reports si no existe
os.makedirs("reports", exist_ok=True)

# 1️⃣ Cargar CSVs desde la carpeta data
df_customers = pd.read_csv("data/customers.csv")
df_orders = pd.read_csv("data/orders.csv")
df_order_items = pd.read_csv("data/order_items.csv")
df_products = pd.read_csv("data/products.csv")

# 2️⃣ Merge de todas las tablas para tener un dataframe completo
df_merged = (
    df_order_items
    .merge(df_orders, on="order_id", how="left")
    .merge(df_products, on="product_id", how="left")
    .merge(df_customers, on="customer_id", how="left")
)

# 3️⃣ KPIs GENERALES
ventas_totales = df_merged["total"].sum()
ticket_medio = df_merged["total"].mean()
num_clientes = df_merged["customer_id"].nunique()
num_pedidos = df_merged["order_id"].nunique()

print("✅ KPIs GENERALES")
print(f"Ventas totales: {ventas_totales} €")
print(f"Ticket medio: {ticket_medio:.2f} €")
print(f"Nº de clientes: {num_clientes}")
print(f"Nº de pedidos: {num_pedidos}")

# 4️⃣ Ventas por categoría
ventas_categoria = df_merged.groupby("category")["total"].sum()
ventas_categoria.to_csv("reports/ventas_por_categoria.csv")
print("\n✅ Ventas por categoría:")
print(ventas_categoria)

# Gráfico de barras por categoría
plt.figure(figsize=(8,5))
sns.barplot(x=ventas_categoria.index, y=ventas_categoria.values, palette="viridis")
plt.title("Ventas por categoría")
plt.ylabel("Ventas (€)")
plt.xlabel("Categoría")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reports/ventas_por_categoria.png")
plt.show()

# 5️⃣ Top productos
top_productos = df_merged.groupby("product_name")["total"].sum().sort_values(ascending=False)
top_productos.to_csv("reports/top_productos.csv")
print("\n✅ Top productos por ventas:")
print(top_productos)

plt.figure(figsize=(10,6))
sns.barplot(x=top_productos.index, y=top_productos.values, palette="magma")
plt.title("Top productos por ventas")
plt.ylabel("Ventas (€)")
plt.xlabel("Producto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reports/top_productos.png")
plt.show()

# 6️⃣ Top clientes
top_clientes = df_merged.groupby("customer_name")["total"].sum().sort_values(ascending=False)
top_clientes.to_csv("reports/top_clientes.csv")
print("\n✅ Top clientes:")
print(top_clientes)

plt.figure(figsize=(8,5))
sns.barplot(x=top_clientes.index, y=top_clientes.values, palette="coolwarm")
plt.title("Top clientes por ventas")
plt.ylabel("Ventas (€)")
plt.xlabel("Cliente")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reports/top_clientes.png")
plt.show()

# 7️⃣ Ventas por ciudad
ventas_ciudad = df_merged.groupby("city")["total"].sum()
ventas_ciudad.to_csv("reports/ventas_por_ciudad.csv")
print("\n✅ Ventas por ciudad:")
print(ventas_ciudad)

plt.figure(figsize=(6,6))
ventas_ciudad.plot.pie(autopct='%1.1f%%', title="Ventas por ciudad")
plt.ylabel("")
plt.tight_layout()
plt.savefig("reports/ventas_por_ciudad.png")
plt.show()

# 8️⃣ Ventas por mes
df_merged['order_date'] = pd.to_datetime(df_merged['order_date'])
ventas_mes = df_merged.groupby(df_merged['order_date'].dt.to_period('M'))['total'].sum()
ventas_mes.to_csv("reports/ventas_por_mes.csv")
print("\n✅ Ventas por mes:")
print(ventas_mes)

plt.figure(figsize=(10,5))
ventas_mes.index = ventas_mes.index.astype(str)
sns.lineplot(x=ventas_mes.index, y=ventas_mes.values, marker='o')
plt.title("Ventas por mes")
plt.ylabel("Ventas (€)")
plt.xlabel("Mes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reports/ventas_por_mes.png")
plt.show()
```

```python
# ==========================================================
# Proyecto Portfolio Profesional: Análisis de Datos
# Autor: Junior FP DAM (con supervisión de analista senior)
# Objetivo: Cargar CSVs, generar df_merged, KPIs, gráficos y exportar resultados
# ==========================================================

# 1️⃣ Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configuración de gráficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# 2️⃣ Definir rutas de carpetas
data_path = os.path.join(os.getcwd(), 'data')
reports_path = os.path.join(os.getcwd(), 'reports')

# Crear carpeta reports si no existe
os.makedirs(reports_path, exist_ok=True)

# 3️⃣ Cargar CSVs
try:
    df_customers = pd.read_csv(os.path.join(data_path, 'customers.csv'))
    df_orders = pd.read_csv(os.path.join(data_path, 'orders.csv'))
    df_order_items = pd.read_csv(os.path.join(data_path, 'order_items.csv'))
    df_products = pd.read_csv(os.path.join(data_path, 'products.csv'))
    print("✅ CSVs cargados correctamente")
except FileNotFoundError as e:
    print(f"❌ Error al cargar CSVs: {e}")

# 4️⃣ Crear df_merged (unión de todas las tablas)
# Se hace merge paso a paso para evitar KeyError
df_merged = df_order_items.merge(df_orders, on='order_id', how='left') \
                          .merge(df_customers, on='customer_id', how='left') \
                          .merge(df_products, on='product_id', how='left')

# Verificar columnas importantes
if 'total' not in df_merged.columns:
    df_merged['total'] = df_merged['qty'] * df_merged['unit_price']

print("✅ df_merged creado con éxito")
print(df_merged.head())

# 5️⃣ Cálculo de KPIs generales
total_ventas = df_merged['total'].sum()
ticket_medio = df_merged.groupby('order_id')['total'].sum().mean()
top_clientes = df_merged.groupby('customer_name')['total'].sum().sort_values(ascending=False).head(10)
top_productos = df_merged.groupby('product_name')['total'].sum().sort_values(ascending=False).head(10)
ventas_por_categoria = df_merged.groupby('category')['total'].sum()
ventas_por_ciudad = df_merged.groupby('city')['total'].sum()
ventas_por_mes = df_merged.groupby(pd.to_datetime(df_merged['order_date']).dt.to_period('M'))['total'].sum()

print(f"Total ventas: {total_ventas}")
print(f"Ticket medio: {ticket_medio}")

# 6️⃣ Exportar resultados a CSV y Excel
df_merged.to_csv(os.path.join(reports_path, 'df_merged.csv'), index=False)
df_merged.to_excel(os.path.join(reports_path, 'df_merged.xlsx'), index=False)
top_clientes.to_csv(os.path.join(reports_path, 'top_clientes.csv'))
top_productos.to_csv(os.path.join(reports_path, 'top_productos.csv'))

print("✅ Resultados exportados a reports/")

# 7️⃣ Generar gráficos reproducibles

# Clientes por ciudad
plt.figure()
ventas_por_ciudad.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Ventas por Ciudad')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(reports_path, 'ventas_por_ciudad.png'))
plt.show()

# Ventas por categoría
plt.figure()
ventas_por_categoria.sort_values(ascending=False).plot(kind='bar', color='salmon')
plt.title('Ventas por Categoría')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(reports_path, 'ventas_por_categoria.png'))
plt.show()

# Top 10 productos
plt.figure()
top_productos.sort_values(ascending=True).plot(kind='barh', color='green')
plt.title('Top 10 Productos')
plt.xlabel('Total Ventas')
plt.tight_layout()
plt.savefig(os.path.join(reports_path, 'top_productos.png'))
plt.show()

# Top 10 clientes
plt.figure()
top_clientes.sort_values(ascending=True).plot(kind='barh', color='purple')
plt.title('Top 10 Clientes')
plt.xlabel('Total Comprado')
plt.tight_layout()
plt.savefig(os.path.join(reports_path, 'top_clientes.png'))
plt.show()

# Ventas por mes
plt.figure()
ventas_por_mes.plot(marker='o')
plt.title('Ventas Mensuales')
plt.ylabel('Total Ventas')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(reports_path, 'ventas_por_mes.png'))
plt.show()

print("✅ Gráficos generados y guardados en reports/")
```

```python
# ==========================================================
# Proyecto Portfolio Profesional: Excel Resumido
# Autor: Junior FP DAM (con supervisión de analista senior)
# Objetivo: Crear Excel resumido con KPIs y gráficos embebidos
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import os
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# 1️⃣ Rutas de carpetas
data_path = os.path.join(os.getcwd(), 'data')
reports_path = os.path.join(os.getcwd(), 'reports')
os.makedirs(reports_path, exist_ok=True)

# 2️⃣ Cargar df_merged
df_merged = pd.read_csv(os.path.join(data_path, 'df_merged.csv'))

# 3️⃣ Calcular KPIs
total_ventas = df_merged['total'].sum()
ticket_medio = df_merged.groupby('order_id')['total'].sum().mean()
top_clientes = df_merged.groupby('customer_name')['total'].sum().sort_values(ascending=False).head(10)
top_productos = df_merged.groupby('product_name')['total'].sum().sort_values(ascending=False).head(10)
ventas_por_categoria = df_merged.groupby('category')['total'].sum()
ventas_por_ciudad = df_merged.groupby('city')['total'].sum()
ventas_por_mes = df_merged.groupby(pd.to_datetime(df_merged['order_date']).dt.to_period('M'))['total'].sum()

# 4️⃣ Guardar gráficos como imágenes para Excel
grafico_rutas = {}

# Ventas por ciudad
plt.figure()
ventas_por_ciudad.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Ventas por Ciudad')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
ciudad_img = os.path.join(reports_path, 'ventas_por_ciudad.png')
plt.savefig(ciudad_img)
plt.close()
grafico_rutas['Ventas por Ciudad'] = ciudad_img

# Ventas por categoría
plt.figure()
ventas_por_categoria.sort_values(ascending=False).plot(kind='bar', color='salmon')
plt.title('Ventas por Categoría')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
categoria_img = os.path.join(reports_path, 'ventas_por_categoria.png')
plt.savefig(categoria_img)
plt.close()
grafico_rutas['Ventas por Categoría'] = categoria_img

# Top 10 productos
plt.figure()
top_productos.sort_values(ascending=True).plot(kind='barh', color='green')
plt.title('Top 10 Productos')
plt.xlabel('Total Ventas')
plt.tight_layout()
productos_img = os.path.join(reports_path, 'top_productos.png')
plt.savefig(productos_img)
plt.close()
grafico_rutas['Top Productos'] = productos_img

# Top 10 clientes
plt.figure()
top_clientes.sort_values(ascending=True).plot(kind='barh', color='purple')
plt.title('Top 10 Clientes')
plt.xlabel('Total Comprado')
plt.tight_layout()
clientes_img = os.path.join(reports_path, 'top_clientes.png')
plt.savefig(clientes_img)
plt.close()
grafico_rutas['Top Clientes'] = clientes_img

# Ventas por mes
plt.figure()
ventas_por_mes.plot(marker='o')
plt.title('Ventas Mensuales')
plt.ylabel('Total Ventas')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.tight_layout()
mes_img = os.path.join(reports_path, 'ventas_por_mes.png')
plt.savefig(mes_img)
plt.close()
grafico_rutas['Ventas por Mes'] = mes_img

# 5️⃣ Crear Excel resumido
excel_resumido = os.path.join(reports_path, 'Resumen_KPIs.xlsx')
with pd.ExcelWriter(excel_resumido, engine='openpyxl') as writer:
    # Hoja resumen con KPIs principales
    resumen = pd.DataFrame({
        'KPI': ['Total Ventas', 'Ticket Medio'],
        'Valor': [total_ventas, ticket_medio]
    })
    resumen.to_excel(writer, sheet_name='Resumen KPIs', index=False)
    
    # Hoja top clientes
    top_clientes_df = top_clientes.reset_index().rename(columns={'customer_name':'Cliente','total':'Total Comprado'})
    top_clientes_df.to_excel(writer, sheet_name='Top Clientes', index=False)
    
    # Hoja top productos
    top_productos_df = top_productos.reset_index().rename(columns={'product_name':'Producto','total':'Total Ventas'})
    top_productos_df.to_excel(writer, sheet_name='Top Productos', index=False)
    
    # Hoja ventas por categoría
    ventas_por_categoria_df = ventas_por_categoria.reset_index().rename(columns={'category':'Categoría','total':'Total Ventas'})
    ventas_por_categoria_df.to_excel(writer, sheet_name='Ventas por Categoría', index=False)
    
    # Hoja ventas por ciudad
    ventas_por_ciudad_df = ventas_por_ciudad.reset_index().rename(columns={'city':'Ciudad','total':'Total Ventas'})
    ventas_por_ciudad_df.to_excel(writer, sheet_name='Ventas por Ciudad', index=False)
    
    # Hoja ventas por mes
    ventas_por_mes_df = ventas_por_mes.reset_index().rename(columns={'order_date':'Mes','total':'Total Ventas'})
    ventas_por_mes_df.to_excel(writer, sheet_name='Ventas por Mes', index=False)

# 6️⃣ Insertar gráficos en el Excel
wb = load_workbook(excel_resumido)

for sheet_name, img_path in grafico_rutas.items():
    if sheet_name in wb.sheetnames:
        ws = wb[sheet_name]
    else:
        ws = wb.create_sheet(sheet_name)
    img = Image(img_path)
    ws.add_image(img, 'D2')  # Posición aproximada, ajustar según conveniencia

wb.save(excel_resumido)
print(f"✅ Excel resumido generado con KPIs y gráficos: {excel_resumido}")
```

```python
# ==========================================================
# Proyecto Portfolio Profesional: Excel Resumido
# Autor: Junior FP DAM (con supervisión de analista senior)
# Objetivo: Crear Excel resumido con KPIs y gráficos embebidos
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import os
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# 1️⃣ Rutas de carpetas
reports_path = os.path.join(os.getcwd(), 'reports')
os.makedirs(reports_path, exist_ok=True)

# 2️⃣ Cargar df_merged desde reports/
df_merged = pd.read_csv(os.path.join(reports_path, 'df_merged.csv'))

# 3️⃣ Calcular KPIs
total_ventas = df_merged['total'].sum()
ticket_medio = df_merged.groupby('order_id')['total'].sum().mean()
top_clientes = df_merged.groupby('customer_name')['total'].sum().sort_values(ascending=False).head(10)
top_productos = df_merged.groupby('product_name')['total'].sum().sort_values(ascending=False).head(10)
ventas_por_categoria = df_merged.groupby('category')['total'].sum()
ventas_por_ciudad = df_merged.groupby('city')['total'].sum()
ventas_por_mes = df_merged.groupby(pd.to_datetime(df_merged['order_date']).dt.to_period('M'))['total'].sum()

# 4️⃣ Guardar gráficos como imágenes para Excel
grafico_rutas = {}

# Ventas por ciudad
plt.figure()
ventas_por_ciudad.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Ventas por Ciudad')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
ciudad_img = os.path.join(reports_path, 'ventas_por_ciudad.png')
plt.savefig(ciudad_img)
plt.close()
grafico_rutas['Ventas por Ciudad'] = ciudad_img

# Ventas por categoría
plt.figure()
ventas_por_categoria.sort_values(ascending=False).plot(kind='bar', color='salmon')
plt.title('Ventas por Categoría')
plt.ylabel('Total Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
categoria_img = os.path.join(reports_path, 'ventas_por_categoria.png')
plt.savefig(categoria_img)
plt.close()
grafico_rutas['Ventas por Categoría'] = categoria_img

# Top 10 productos
plt.figure()
top_productos.sort_values(ascending=True).plot(kind='barh', color='green')
plt.title('Top 10 Productos')
plt.xlabel('Total Ventas')
plt.tight_layout()
productos_img = os.path.join(reports_path, 'top_productos.png')
plt.savefig(productos_img)
plt.close()
grafico_rutas['Top Productos'] = productos_img

# Top 10 clientes
plt.figure()
top_clientes.sort_values(ascending=True).plot(kind='barh', color='purple')
plt.title('Top 10 Clientes')
plt.xlabel('Total Comprado')
plt.tight_layout()
clientes_img = os.path.join(reports_path, 'top_clientes.png')
plt.savefig(clientes_img)
plt.close()
grafico_rutas['Top Clientes'] = clientes_img

# Ventas por mes
plt.figure()
ventas_por_mes.plot(marker='o')
plt.title('Ventas Mensuales')
plt.ylabel('Total Ventas')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.tight_layout()
mes_img = os.path.join(reports_path, 'ventas_por_mes.png')
plt.savefig(mes_img)
plt.close()
grafico_rutas['Ventas por Mes'] = mes_img

# 5️⃣ Crear Excel resumido
excel_resumido = os.path.join(reports_path, 'Resumen_KPIs.xlsx')
with pd.ExcelWriter(excel_resumido, engine='openpyxl') as writer:
    # Hoja resumen con KPIs principales
    resumen = pd.DataFrame({
        'KPI': ['Total Ventas', 'Ticket Medio'],
        'Valor': [total_ventas, ticket_medio]
    })
    resumen.to_excel(writer, sheet_name='Resumen KPIs', index=False)
    
    # Hoja top clientes
    top_clientes_df = top_clientes.reset_index().rename(columns={'customer_name':'Cliente','total':'Total Comprado'})
    top_clientes_df.to_excel(writer, sheet_name='Top Clientes', index=False)
    
    # Hoja top productos
    top_productos_df = top_productos.reset_index().rename(columns={'product_name':'Producto','total':'Total Ventas'})
    top_productos_df.to_excel(writer, sheet_name='Top Productos', index=False)
    
    # Hoja ventas por categoría
    ventas_por_categoria_df = ventas_por_categoria.reset_index().rename(columns={'category':'Categoría','total':'Total Ventas'})
    ventas_por_categoria_df.to_excel(writer, sheet_name='Ventas por Categoría', index=False)
    
    # Hoja ventas por ciudad
    ventas_por_ciudad_df = ventas_por_ciudad.reset_index().rename(columns={'city':'Ciudad','total':'Total Ventas'})
    ventas_por_ciudad_df.to_excel(writer, sheet_name='Ventas por Ciudad', index=False)
    
    # Hoja ventas por mes
    ventas_por_mes_df = ventas_por_mes.reset_index().rename(columns={'order_date':'Mes','total':'Total Ventas'})
    ventas_por_mes_df.to_excel(writer, sheet_name='Ventas por Mes', index=False)

# 6️⃣ Insertar gráficos en el Excel
wb = load_workbook(excel_resumido)

for sheet_name, img_path in grafico_rutas.items():
    if sheet_name in wb.sheetnames:
        ws = wb[sheet_name]
    else:
        ws = wb.create_sheet(sheet_name)
    img = Image(img_path)
    ws.add_image(img, 'D2')  # Posición aproximada, se puede ajustar

wb.save(excel_resumido)
print(f"✅ Excel resumido generado con KPIs y gráficos en: {excel_resumido}")
```

```python
# ==========================================================
# Proyecto Portfolio Profesional: Datos para Tableau
# Autor: Junior FP DAM (con supervisión de analista senior)
# Objetivo: Crear CSVs limpios para dashboards interactivos
# ==========================================================

import pandas as pd
import os

# 1️⃣ Definir rutas
reports_path = os.path.join(os.getcwd(), 'reports')
tableau_path = os.path.join(reports_path, 'tableau')

# Crear carpeta tableau si no existe
os.makedirs(tableau_path, exist_ok=True)

# 2️⃣ Cargar df_merged desde reports
df_merged = pd.read_csv(os.path.join(reports_path, 'df_merged.csv'))

# Convertir la fecha a tipo datetime por seguridad
df_merged['order_date'] = pd.to_datetime(df_merged['order_date'], errors='coerce')

# 3️⃣ Crear datasets limpios para Tableau

# Ventas por mes
ventas_mes = df_merged.groupby(df_merged['order_date'].dt.to_period('M'))['total'].sum().reset_index()
ventas_mes['order_date'] = ventas_mes['order_date'].astype(str)

# Top productos (top 20)
top_productos = df_merged.groupby('product_name')['total'].sum().sort_values(ascending=False).head(20).reset_index()

# Top clientes (top 20)
top_clientes = df_merged.groupby('customer_name')['total'].sum().sort_values(ascending=False).head(20).reset_index()

# Ventas por categoría
ventas_categoria = df_merged.groupby('category')['total'].sum().reset_index()

# Ventas por ciudad
ventas_ciudad = df_merged.groupby('city')['total'].sum().reset_index()

# Dataset completo limpio (por si Tableau lo necesita)
df_completo = df_merged.copy()

# 4️⃣ Exportar los CSVs a /reports/tableau/
ventas_mes.to_csv(os.path.join(tableau_path, 'ventas_por_mes.csv'), index=False)
top_productos.to_csv(os.path.join(tableau_path, 'top_productos.csv'), index=False)
top_clientes.to_csv(os.path.join(tableau_path, 'top_clientes.csv'), index=False)
ventas_categoria.to_csv(os.path.join(tableau_path, 'ventas_por_categoria.csv'), index=False)
ventas_ciudad.to_csv(os.path.join(tableau_path, 'ventas_por_ciudad.csv'), index=False)
df_completo.to_csv(os.path.join(tableau_path, 'df_completo_para_tableau.csv'), index=False)

print("✅ CSVs creados correctamente en /reports/tableau")
print("Archivos generados:")
print("- ventas_por_mes.csv")
print("- top_productos.csv")
print("- top_clientes.csv")
print("- ventas_por_categoria.csv")
print("- ventas_por_ciudad.csv")
print("- df_completo_para_tableau.csv")
```

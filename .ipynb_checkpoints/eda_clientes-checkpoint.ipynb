```python
# EDA INICIAL: CLIENTES


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# 1️⃣ Cargar Excel desde outputs
df = pd.read_excel("outputs/summary_analysis.xlsx")

# 2️⃣ Revisar primeras filas y estructura
print("Primeras filas:")
display(df.head())

print("\nInformación del DataFrame:")
df.info()

# 3️⃣ Revisar nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# 4️⃣ Revisar columnas y tipos
print("\nColumnas disponibles:")
print(df.columns)

# 5️⃣ Estadísticas básicas (solo para columnas numéricas, si hay)
print("\nEstadísticas descriptivas:")
display(df.describe())

# 6️⃣ Cantidad de clientes por ciudad
clientes_por_ciudad = df['city'].value_counts()
print("\nCantidad de clientes por ciudad:")
print(clientes_por_ciudad)

# 7️⃣ Gráfico: clientes por ciudad
plt.figure(figsize=(10,6))
sns.barplot(x=clientes_por_ciudad.index, y=clientes_por_ciudad.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Número de clientes por ciudad")
plt.ylabel("Cantidad de clientes")
plt.xlabel("Ciudad")
plt.show()

# 8️⃣ Listado de clientes únicos
clientes_unicos = df['customer_name'].nunique()
print(f"\nNúmero de clientes únicos: {clientes_unicos}")
```

```python
# EDA AVANZADO: CLIENTES + ÓRDENES + PRODUCTOS

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# 1️⃣ Cargar los datos
df_customers = pd.read_excel("summary_analysis.xlsx")      # Clientes
df_orders = pd.read_csv("orders.csv")                      # Órdenes
df_order_items = pd.read_csv("order_items.csv")            # Detalle de órdenes
df_products = pd.read_csv("products.csv")                  # Productos

# Revisar primeras filas
display(df_customers.head())
display(df_orders.head())
display(df_order_items.head())
display(df_products.head())

# 2️⃣ Unir datos: order_items + orders + customers + products
df_full = df_order_items.merge(df_orders, on="order_id", how="left") \
                        .merge(df_customers, on="customer_id", how="left") \
                        .merge(df_products, on="product_id", how="left")

# Añadir columna total por línea (qty * unit_price)
df_full['total'] = df_full['qty'] * df_full['unit_price']

# Revisar DataFrame completo
display(df_full.head())

# 3️⃣ Análisis agregados

# 3a. Total por cliente
total_por_cliente = df_full.groupby('customer_name')['total'].sum().sort_values(ascending=False)
print("Total gastado por cliente:")
display(total_por_cliente)

# 3b. Total por ciudad
total_por_ciudad = df_full.groupby('city')['total'].sum().sort_values(ascending=False)
print("Total gastado por ciudad:")
display(total_por_ciudad)

# 3c. Cantidad de órdenes por cliente
ordenes_por_cliente = df_full.groupby('customer_name')['order_id'].nunique().sort_values(ascending=False)
print("Cantidad de órdenes por cliente:")
display(ordenes_por_cliente)

# 4️⃣ Gráficos automáticos

# 4a. Clientes por gasto total
plt.figure(figsize=(10,6))
sns.barplot(x=total_por_cliente.index, y=total_por_cliente.values, palette="viridis")
plt.xticks(rotation=45)
plt.ylabel("Total gastado")
plt.title("Total gastado por cliente")
plt.show()

# 4b. Total gastado por ciudad
plt.figure(figsize=(10,6))
sns.barplot(x=total_por_ciudad.index, y=total_por_ciudad.values, palette="magma")
plt.xticks(rotation=45)
plt.ylabel("Total gastado")
plt.title("Total gastado por ciudad")
plt.show()

# 4c. Cantidad de órdenes por cliente
plt.figure(figsize=(10,6))
sns.barplot(x=ordenes_por_cliente.index, y=ordenes_por_cliente.values, palette="cividis")
plt.xticks(rotation=45)
plt.ylabel("Cantidad de órdenes")
plt.title("Órdenes por cliente")
plt.show()

# 5️⃣ Exportar resúmenes a outputs
total_por_cliente.to_csv("outputs/total_por_cliente.csv")
total_por_ciudad.to_csv("outputs/total_por_ciudad.csv")
ordenes_por_cliente.to_csv("outputs/ordenes_por_cliente.csv")

print("✅ Análisis completo generado y resúmenes exportados a 'outputs/'.")
```

```python
import pandas as pd

# 1. Cargar CSVs desde la carpeta data/
df_customers = pd.read_csv("data/customers.csv")
df_orders = pd.read_csv("data/orders.csv")
df_order_items = pd.read_csv("data/order_items.csv")
df_products = pd.read_csv("data/products.csv")

print("✅ Archivos cargados correctamente\n")

print("Customers:")
display(df_customers.head())

print("\nOrders:")
display(df_orders.head())

print("\nOrder Items:")
display(df_order_items.head())

print("\nProducts:")
display(df_products.head())
```

```python
import pandas as pd

# 1. Crear columna total
df_order_items["total"] = df_order_items["qty"] * df_order_items["unit_price"]

# 2. Merge con productos
df_items_prod = df_order_items.merge(df_products, on="product_id", how="left")

# 3. Merge con órdenes
df_items_prod_orders = df_items_prod.merge(df_orders, on="order_id", how="left")

# 4. Merge con clientes
df_full = df_items_prod_orders.merge(df_customers, on="customer_id", how="left")

print("✅ Tabla unificada lista")
display(df_full.head())
print("\nColumnas:", df_full.columns)
```

```python
import pandas as pd

# --- KPIs GENERALES ---
ventas_totales = df_full["total"].sum()
ticket_medio = df_full.groupby("order_id")["total"].sum().mean()
num_clientes = df_full["customer_id"].nunique()
num_pedidos = df_full["order_id"].nunique()

print("✅ KPIs GENERALES")
print(f"Ventas totales: {ventas_totales} €")
print(f"Ticket medio: {ticket_medio:.2f} €")
print(f"Nº de clientes: {num_clientes}")
print(f"Nº de pedidos: {num_pedidos}")

# --- VENTAS POR CATEGORÍA ---
ventas_categoria = df_full.groupby("category")["total"].sum().sort_values(ascending=False)
print("\n✅ Ventas por categoría:")
print(ventas_categoria)

# --- TOP PRODUCTOS ---
top_productos = df_full.groupby("product_name")["total"].sum().sort_values(ascending=False)
print("\n✅ Top productos por ventas:")
print(top_productos)

# --- TOP CLIENTES ---
top_clientes = df_full.groupby("customer_name")["total"].sum().sort_values(ascending=False)
print("\n✅ Top clientes:")
print(top_clientes)

# --- VENTAS POR CIUDAD ---
ventas_ciudad = df_full.groupby("city")["total"].sum().sort_values(ascending=False)
print("\n✅ Ventas por ciudad:")
print(ventas_ciudad)

# --- VENTAS POR MES ---
df_full["order_date"] = pd.to_datetime(df_full["order_date"])
ventas_mes = df_full.groupby(df_full["order_date"].dt.to_period("M"))["total"].sum()
print("\n✅ Ventas por mes:")
print(ventas_mes)
```
